# Detecting Overfitting and underfitting with learning curves
here, I have analysed the learning curves of the three models namely Decision tree, logistic regression and support vector machines. 



### Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
```

### Import the dataset

```python
data = pd.read_csv('data.csv')
X = np.array(data[['x1', 'x2']])
y = np.array(data['y'])
```

### Import the models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
```

### function for drawing curves

```python
def draw_learning_curves(modname, X, y, estimator, num_trainings):
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=None, n_jobs=1, train_sizes=np.linspace(.1, 1.0, num_trainings))
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)

    plt.grid()

    plt.title(modname + ": Learning Curves")
    plt.xlabel("Training examples")
    plt.ylabel("Score")

    plt.plot(train_scores_mean, 'o-', color="g", label="Training score")
    plt.plot(test_scores_mean, 'o-', color="y", label="Cross-validation score")

    plt.legend(loc="best")
    plt.show()
```

### creating the models

```python
logReg_md = LogisticRegression()
GradBoost_md = GradientBoostingClassifier()
SVC_md = SVC(kernel = 'rbf', gamma = 1000)
```

### drawing the curves

```python
draw_learning_curves("Logistic Regression", X,y, logReg_md, 12)
draw_learning_curves("Decision Tree", X,y, GradBoost_md, 12)
draw_learning_curves("SVC", X,y, SVC_md, 12)
```

### Conclusion
For this dataset,
Decision Tree with Gradient Boosting is the best model whereas Logistic Regression underfits the data as the model is converging but the training error is more and Support Vector machine overfits the data as the model doesn't converge.
